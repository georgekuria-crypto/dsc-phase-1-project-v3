```python
#picking out the columns that will be relevant to solve this problem
utilized_columns=[
'Event.Id' , 'Event.Date' , 'Location', 'Country', 'Airport.Name', 'Injury.Severity', 'Aircraft.damage', 'Aircraft.Category',
'Make', 'Model' , 'Number.of.Engines', 'Engine.Type' ,  'Purpose.of.flight' , 'Total.Fatal.Injuries' , 'Total.Serious.Injuries', 'Total.Minor.Injuries', 'Total.Uninjured',
'Weather.Condition'
]
#making my new dataframe consist of only the selected columns
df=df[utilized_columns]

df
```

```python
#checking for the unique values in the new relevant dataframe for our problem
unique_values= df.nunique()

unique_values
```

```python
#checking the data types of the selected columns
df.dtypes
```

```python
#converting the event_date column to date time type
df['Event.Date'] = pd.to_datetime(df['Event.Date'], errors='coerce')
```

```python
#converting the data type of the columns Total fatal injuries,total serious injuries,total minor injuries and total uninjured into integer type
for value in ['Total.Fatal.Injuries', 'Total.Serious.Injuries', 'Total.Minor.Injuries','Total.Uninjured']:
    df[value] = pd.to_numeric(df[value], errors='coerce').fillna(0).astype(int)

df
```

```python
#confirmation of the made changes 
df.dtypes
```

```python
#checking the sum of the missing values in each column
df.isna().sum()
```

```python
#identifying columns with missing values in terms of percentages
missing = df.isnull().sum().sort_values(ascending=False)

missing_percent = (missing / len(df)) * 100

print(f'Missing values (%): {missing_percent}')
```

```python
#dropping the Aircraft category and Airport name columns due to the high amounts of missing data
df=df.drop(['Aircraft.Category','Airport.Name'],axis=1)
```

```python
#dropping missing values in the Engine.Type,Purpose.of.flight ,Number.of.Engines,Weather.Condition,Aircraft.damage,Injury.Severity,Country,Model,Make,Location,Event.Date,Event.Id columns

df.dropna(subset=['Engine.Type','Purpose.of.flight','Number.of.Engines','Weather.Condition','Aircraft.damage','Injury.Severity','Country','Model','Make','Location','Event.Date','Event.Id'], inplace=True)
```

```python
#checking the data types of each column and the number of rows after the removal of null values
df.info()
```

```python
#Checking that there are no missing values  
df.isnull().sum()
```

```python
#Checking for duplicates
df.duplicated().sum()
```

```python
#Removing duplicates based on columns
df=df.drop_duplicates()
```

```python
#Confirming that there are no duplicates
df.duplicated().sum()
```

```python
 #Calculate Fatality rate and create a new column ( Fatality rate = [Total Fatal Injuries] / [Total Passengers] )

df['Total.Passengers'] = (
    df['Total.Fatal.Injuries'] +
    df['Total.Serious.Injuries'] +
    df['Total.Minor.Injuries'] +
    df['Total.Uninjured']
)

# Calculate Fatality Rate
df['Fatality.Rate'] = df['Total.Fatal.Injuries'] / df['Total.Passengers']
```

```python
df
```

```python
#Determining the different weather conditions in the dataframe
df['Weather.Condition'].value_counts()
```

```python
#Creating a new well defined column specifying the type of weather and its effect on the navigation of the aircrafts
#Dictionary created redefining the strings in the weather.condition column
weather_conditions={'VMC':'Visible Weather Conditions','IMC':'Instrument Required Weather Conditions','UNK':' Slightly Unknown Weather Conditions','Unk':'Strongly Unknown Weather Conditions'}

#Mapping the dictionary to the weather.condition column and creating a new column
df['Weather.Recorded']=df['Weather.Condition'].map(weather_conditions)

df
```

```python
df.to_csv("Clean_Aviation_Data.csv",index=False)
```

```python
dangerous_aircraft_make = df.groupby('Make')['Total.Fatal.Injuries'].mean()

# Sorted Display of Fatalities according to aircraft make
top_dangerous_aircraft_make = dangerous_aircraft_make.sort_values(ascending=False).head(10)

colors = ['red', 'orange', 'yellow', 'green', 'blue', 'indigo', 'violet', 'cyan', 'magenta', 'gray']

plt.figure(figsize=(12, 6))
top_dangerous_aircraft_make.plot(kind='bar', color=colors)
plt.ylabel('Average No. of Total Fatal Injuries')
plt.xlabel('Aircraft Make')
plt.title('The Top 10 Aircraft Makes With Highest Average Fatalities')
plt.xticks(rotation=45, ha='right')#sets the position of the labels on the x-axis
plt.tight_layout()
plt.show()
```

```python
dangerous_aircraft_model = df.groupby('Model')['Total.Fatal.Injuries'].mean()

# Sorted Display of Fatalities according to aircraft make
top_dangerous_aircraft_model = dangerous_aircraft_model.sort_values(ascending=False).head(10)

plt.figure(figsize=(12, 6))
top_dangerous_aircraft_model.plot(kind='bar', color='purple')
plt.ylabel('Average No. of Total Fatal Injuries')
plt.xlabel('Aircraft Model')
plt.title('The Top 10 Aircraft Models With Highest Average Fatalities')
plt.xticks(rotation=45, ha='right')#sets the position of the labels on the x-axis
plt.tight_layout()
plt.show()
```
